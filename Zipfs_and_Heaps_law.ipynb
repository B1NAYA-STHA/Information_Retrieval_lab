```python
import nltk
import pandas as pd
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
from nltk import FreqDist
import matplotlib.pyplot as plt

nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('punkt_tab')
```

**Zipf's Law**

```python
#frequency_distribution

tokens = word_tokenize(gutenberg.raw('austen-emma.txt'))
words = ([token.lower() for token in tokens if token.isalpha()])
freq_dist = FreqDist(words)
freq_dist
```

```python
#Frequency and ranks

freqs = sorted(freq_dist.values(), reverse=True)
ranks = range(1, len(freqs)+ 1)
```

```python
ranks
```

```python
# Plot Zipf's Law
plt.figure(figsize=(10, 5))
plt.loglog(ranks, freqs)
plt.title("Zipf's Law: Word Frequency vs. Rank")
plt.xlabel("Rank of word")
plt.ylabel("Frequency of word")
plt.grid(True)
plt.show()
```

**Heap's Law**

```python
token = words  # From Zipf example
vocab = set()
token_counts = []
vocab_sizes = []

step = 1000
for i in range(step, len(token), step):
    subset = token[:i]
    vocab.update(subset)
    token_counts.append(i)
    vocab_sizes.append(len(vocab))
```

```python
# Plot Heaps' Law
plt.figure(figsize=(10, 5))
plt.plot(token_counts, vocab_sizes)
plt.title("Heaps' Law: Vocabulary Growth")
plt.xlabel("Total number of tokens")
plt.ylabel("Vocabulary size")
plt.grid(True)
plt.show()
```
